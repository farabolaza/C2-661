## Tipos de variables
- Numéricas:
    - int, float, long, complex
- string
- booleanas
    - True / False

Se resumen en la siguiente tabla

**<center>Python Scalar Types</center>**

| Type        | Example        | Description                                                  |
|-------------|----------------|--------------------------------------------------------------|
| ``str``     | ``x = 'abc'``  | String: caracteres de texto                                   |
| ``int``     | ``x = 1``      | integers (i.e., números enteros)                               |
| ``float``   | ``x = 1.0``    | floating-point numbers (i.e., números reales)                  |

Se usa la función `type()` para conocer el tipo de una variable

## Operadores de cadenas de texto
Los operadores matemáticos son diferentes en las cadenas de teto, por ejemplo el sigo + se usa para unir cadenas

Repeticiones: Se usa *n para repetir una cadena n veces

````python
# Asignar y combinar cadenas
a = 'Hello'
b = 'World'
print(a + ' ' + b)
````

````python
# f-strings: reemplazarvalores de variables en cadenas
firstname = 'Beto'
print(f'Me llamo {firstname}')
````

````python
# Combininar f-strings
firstname = 'Beto'
lastname = 'Velez'

print(f'My name is {firstname} {lastname}')
````

````python
# Repetir una cadena
print(str(a)*5)
````

````python
# Repetir un cadena con espacios
print(str(a+' ')*5)
````

## Operadores Matemáticos
- +, -, *, y /
- Exponenciación **
- Modulo % (Resto)

````python
# Adición
a = 2 + 7
print(f'a = {a}')
````

````python
# Sustracción
b = a - 5
print(f'b = a-5 = {a}-5 = {b}')
````

````python
# Multiplicación
print(f'b*2 = {b}*2 = {b*2}')
````

````python
# División
print(f'a/3 = {a}/3 = {a/3}')
````

````python
# Exponenciación
print(f'b^2 = b**2 = {b}**2 = {b**2}')
````

````python
# Modulo
print('(Modulo de vuelve el resto luego de dividir entre dos números)')
print(f'el resto de  b/9 = resto de {b}/9 = {b%9}')
````

## Operadores lógicos

Hacen una comparación y devuelven un booleano (verdadero o falso)

```
x == y # x es igual a y
x != y # x no es igual a y
x > y # x es mayor que y
x < y # x es menor que y
x >= y # x es mayor o igual que y
x <= y # x es menor o igual que y
```

````python
# En cadenas de texto
a = 'hola'
b = 'que tal'
c = 'Hola'
print('a==b: ' + str(a==b))
print('a==c: ' + str(a==c))
print('a!=b: ' + str(a!=b))
print()
````

````python
# En datos numéricos
d = 5
e = 8
print('d < e: ' + str(d < e))
print('d >= e: ' + str(d >= e))
````
