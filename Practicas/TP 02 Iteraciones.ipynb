## 1. Divisores de un número

### Planteo
Elaborar un programa que solicite a un usuario que ingrese un número entero y muestre todos los divisores de ese número (un número al dividirse por sus divisores arroja resto 0).

### Solución
Una opción es usar un bucle while y un contador junto al operador módulo. Si el modulo de dividir al número ingresado con el contador es cero, signiica que el contador es divisor del número. Incrementando el contador desde 1 hasta el número ingresado se usa un condicional para imprimir sólo aquellos resultados donde num%div==0

`````python
num = int(input("ingrese un número entero: "))
div = 1
while div < num:
  if num%div == 0:
    print(div, "es divisor de ",num)
  div = div+1
`````

Otra opción es usar un bucle for (se pueden almacenar los divisores en una lista usando *append*)

`````python
num = int(input("ingrese un número entero: "))
divisores = []

for i in range(num):
  if num%(i+1) == 0: # uso i+1 porque range va de 0 a n-1. range(4) = 0,1,2,3
    print(i+1, "es divisor de ",num)
    divisores.append(i+1)
if len(divisores) == 2: # si la lista de divsores tiene dos elementos, el número es primo
  print (num, "es primo")
else:
  print(num, "no es primo")
print("divores de ",num, ": " ,divisores)
`````

## 7. Sucesión de Fibonacci

### Planteo
Realizar un programa que imprima los primeros N términos de una sucesión de fibonacci, la que empieza con 0 y 1, y cada término que sigue a estos valores es la suma de los dos valores anteriores. Imprimir la razón entre un número y su número anterior

### Solución

Una aproximación poco eficiente es calcular todo el árbol cada vez que se nos pide un término: si se nos pide el término "n" se calcular todos los anteriores para luego hacer la suma y obtener el que se nos pide. Esto implica que vamos "Bajando" por un proceso en donde a fin de cuentas sólo calcularemos muchas veces $f(0)$ y $f(1)$

matemáticamente tenemos que:

\begin{align}
f_0 &= 0 \\
f_1 &= 1 \\
f_n &= f_{n-1} + f_{n-2} \\
f_2 &= f_{2-1} + f_{2-2} \\
f_2 &= f_{1} + f_{0} \\
f_2 &= 1 + 0 = 1 \\
\end{align}

Calcular $f(3)$ implica calcular $f(2)$ (que incluye a los términos anteriores) y así sucesivamente.

`````python
n = 2 #@param {type:"integer"}
contador = 0 # el contador controla el bucle
# definimos los valores iniciales
f0 = 0  # valor fijo al comienzo de la secuencia
f1 = 1  # valor fijo al comienzo de la secuencia
fn = f0+f1  # termino n de la secuencia

while contador <= n:
  if n <=2:
    if n == 0:
      print(f0)
    if n == 1:
      print(f0,f1)
    if n == 2:
      print(f0,f1,fn)
    break
  else:
    f0 = f1
    f1 = fn
    fn = f0+f1
    contador = contador+1
    print(fn)
`````

`````python
def fibo(n):
  serie = [0,1]
  ratio = ["-",1]
  if n <= 2:
    print(serie[0:n]) # imprime desde las posiciones 0 a la n de la lista serie
    print(ratio[0:n]) # imprime desde las posiciones 0 a la n de la lista ratio
  else:

    for i in range(1,n):
      serie.append(serie[i]+serie[i-1])
      ratio.append(serie[i+1]/serie[i])

    print(serie)
    print(ratio)
`````

Se puede definir un método recurrente (que se llama a sí mismo dentro de su definición) para calcular la sucesión de Fibonacci. Lo pueden googlear, pero es didácticamente basatante oscuro, por lo que no lo recomiendo como material de enseñanza.

Lo que si se puede hacer, es la versión mas sencilla, que también es un poco críptica pero más clara, que sería algo así:

`````python
def fib(n):
    f_n = 0  # posición n
    fn_1 = 1  # posición n+1

    for k in range(n): # recorro de 0 a n calculando los valores
        fn_2 = fn_1+ f_n # posición n+2, para n=0  fn_2 => vale 1=1+0
        f_n = fn_1 # ahora la posición n, que es la que se va a imprimir, se redefine como la posición siguiente f_n=1
        fn_1 = fn_2 # la posiciión n+1 se define como lo que se obtuvo para n+2 y volvemos al inicio del bucle fn_1 = 1

        # para n=1
        # fn_2 = 1+1
        # f_n = 1
        # f_n1 =2

        # para n=2
        # fn_2 = 2+1 =3
        # f_n = 2
        #fn_1 = 3

        # para n=3
        # fn_2 = 3+2= 5
        # fn = 3
        # fn_1 = 3

    return f_n
`````

`````python
for i in range(8):
  print(fib(i))
`````
